```python
url = "https://realpython.github.io/fake-jobs/"
text = requests.get(url).content
```

```python
job_soup = BeautifulSoup(text, 'html.parser')
```

```python
print(job_soup.prettify())
```

```python
# Python iş ilanlarını içeren div elemanlarını bul
python_job_divs = [job_div for job_div in job_soup.find_all('div', class_='card-content') if 'python' in job_div.text.lower()]

# Her bir Python iş ilanının başlığını ve bağlantısını ekrana yazdır
for python_job_div in python_job_divs:
    title = python_job_div.find('h2', class_='title').text.strip()
    link = python_job_div.find('a')['href']
    print(f"Başlık: {title}\nBağlantı: {link}\n---")
```

```python
url = 'https://realpython.github.io/fake-jobs/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]
plt.figure(figsize=(10, 6))
plt.bar(range(len(python_headlines)), word_counts)
plt.xlabel('Başlık')
plt.ylabel('Kelime Sayısı')
plt.title('Python İçeren Başlıkların Kelime Sayısı')
plt.xticks(range(len(python_headlines)), python_headlines, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
url = 'https://realpython.github.io/fake-jobs/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]
plt.figure(figsize=(10, 6))
plt.pie(word_counts, labels=python_headlines, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Python İçeren Başlıkların Kelime Sayısı')
plt.show()
```

```python
import seaborn as sns
from bs4 import BeautifulSoup
import requests

url = 'https://realpython.github.io/fake-jobs/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]

plt.figure(figsize=(10, 6))
sns.violinplot(y=word_counts, color='skyblue')
plt.ylabel('Kelime Sayısı')
plt.title('Python İçeren Başlıklardaki Kelime Sayısının Violin Plot ile Dağılımı')
plt.grid(True)
plt.show()
```
