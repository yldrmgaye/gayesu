```python
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
# Grafiklerin Matrislere dönüştürülmesi
camasirsertligi=[["sert",[0,20,40],[1,1,0]],
                 ["h.sert",[20,40,60],[0,1,0]],
                 ["h.yumuşak",[40,60,80],[0,1,0]],
                 ["yumuşak",[60,80,100],[0,1,1]]]

camasirmiktari=[["az",[0,1,3],[1,1,0]],
                ["normal",[1,3,5],[0,1,0]],
                ["çok",[3,5,6],[0,1,1]]]

devir=         [["hassas",[0,400,600],[1,1,0]],
                ["hafif",[400,600,800],[0,1,0]],
                ["normal",[600,800,1000],[0,1,0]],
                ["güçlü",[800,1000,1200],[0,1,1]]]

aranan_sertlik = 65
aranan_miktar = 4
```

```python
#üyelik fonk. grafiklerinin çizilmesi

fig, ax = plt.subplots() 
ax.set_title('Çamaşır Sertliği Verileri')
ax.plot(camasirsertligi[0][1],camasirsertligi[0][2],label="sert", color="g")
ax.plot(camasirsertligi[1][1],camasirsertligi[1][2],label="h.sert", color="r")
ax.plot(camasirsertligi[2][1],camasirsertligi[2][2],label="h.yumuşak", color="b")
ax.plot(camasirsertligi[3][1],camasirsertligi[3][2],label="yumuşak",color="k")
ax.legend()
ax.scatter(aranan_sertlik, 0.25, color='orange', label='sertlik değeri')

fig, ax = plt.subplots() 
ax.set_title('Çamaşır Miktari Verileri')
ax.plot(camasirmiktari[0][1],camasirmiktari[0][2],label="az",color="g")
ax.plot(camasirmiktari[1][1],camasirmiktari[1][2],label="normal",color="r")
ax.plot(camasirmiktari[2][1],camasirmiktari[2][2],label="çok",color="b")
ax.legend()
ax.scatter(aranan_miktar, 0.5, color='orange', label='miktar')
fig, ax = plt.subplots() 
ax.set_title('Çamaşır Devir Verileri')
ax.plot(devir[0][1],devir[0][2],label="hassas",color="g")
ax.plot(devir[1][1],devir[1][2],label="hafif",color="r")
ax.plot(devir[2][1],devir[2][2],label="normal",color="b")
ax.plot(devir[3][1],devir[3][2],label="güçlü",color="k")
ax.legend()


plt.show()
```

```python
# sertlik üyelikleri fonk.
def camasirsertlikuye(camasirsertligi,aranan_sertlik):
    uyelikler = list()

    if aranan_sertlik >= camasirsertligi[0][1][0] and aranan_sertlik <= camasirsertligi[0][1][2]:
        uyelikler.append(camasirsertligi[0][0])
        if aranan_sertlik >= camasirsertligi[0][1][0] and aranan_sertlik <= camasirsertligi[0][1][1]:
            uyelikler.append(1)
        else:
            x = 1 - ((1-0)*(camasirsertligi[0][1][2]-aranan_sertlik))/(camasirsertligi[0][1][2]-camasirsertligi[0][1][0])
            uyelikler.append(x)
    elif aranan_sertlik >= camasirsertligi[1][1][0] and aranan_sertlik <= camasirsertligi[1][1][2]:
       uyelikler.append(camasirsertligi[1][0])
       if aranan_sertlik >= camasirsertligi[1][1][0] and aranan_sertlik <= camasirsertligi[1][1][1]:
           x=1-((1-0)*(camasirsertligi[1][1][0]-aranan_sertlik))/(camasirsertligi[1][1][0]-camasirsertligi[1][1][1])
           uyelikler.append(x) 
       else:
           x=1-((1-0)*(camasirsertligi[1][1][2]-aranan_sertlik))/(camasirsertligi[1][1][2]-camasirsertligi[1][1][1])
           uyelikler.append(x)
    elif aranan_sertlik >= camasirsertligi[2][1][0] and aranan_sertlik <= camasirsertligi[2][1][2]:
       uyelikler.append(camasirsertligi[2][0])
       if aranan_sertlik >= camasirsertligi[2][1][0] and aranan_sertlik <= camasirsertligi[2][1][1]:
           x=1-((1-0)*(camasirsertligi[2][1][0]-aranan_sertlik))/(camasirsertligi[2][1][0]-camasirsertligi[2][1][1])
           uyelikler.append(x) 
       else:
           x=1-((1-0)*(camasirsertligi[2][1][2]-aranan_sertlik))/(camasirsertligi[2][1][2]-camasirsertligi[2][1][1])
           uyelikler.append(x)
    elif aranan_sertlik >= camasirsertligi[3][1][0] and aranan_sertlik <= camasirsertligi[3][1][2]:
       uyelikler.append(camasirsertligi[3][0])
       if aranan_sertlik >= camasirsertligi[3][1][0] and aranan_sertlik <= camasirsertligi[3][1][1]:
           x=1-((1-0)*(camasirsertligi[3][1][0]-aranan_sertlik))/(camasirsertligi[3][1][0]-camasirsertligi[3][1][1])
           uyelikler.append(x) 
       else:
           uyelikler.append(1)
    return uyelikler
```
